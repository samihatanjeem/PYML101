```python
#Importing pandas and loading data into a DataFrame
import pandas as pd

#LOAD
df = pd.read_csv("Stocks_Tanjeem.csv", index_col="Symbol")
print(df)

#Displaying non-null values for each column and column’s type
print("\nNon-null values per column:")
print(df.count())
print("\nData column types:")
print(df.dtypes)

#Finding the number of rows and columns
print("\nNumber of rows and columns:")
print("Shape:" , df.shape)
print("Number of rows:", len(df))
print("Number of columns:", df.shape[1])  

#Display the first and last 10 rows
print(df.head(10))
print(df.tail(10))
```

```python
#ACCESS
#Display the Company column
print("\nCompany column:")
print(df["Company"])

#Display the columns named Company and Industry
print("\nCompany and Industry columns:")
print(df[["Company", "Industry"]])

#Display the row that has the symbol AAPL
print("\nRows with AAPL:")
print(df.loc["AAPL"])

#Display the Exchange for the row with the Symbol AAPL
print(("\nExchange for the row with AAPL:"))
print(df.loc["AAPL", "Exchange"])
```

```python
#FILTER
#Sorting data by Industry andCompany
sorted_data_df= df.sort_values(by=["Industry", "Company"])
print("\nSorted data by Industry and Company:")
print(sorted_data_df)

#Display the rows where Industry is Information Technology
info_tech_df= df[df["Industry"] == "Information technology"]
print("\nRows where Industry is Information Technology:")
print(info_tech_df)
```

```python
#MISSING
#Display the rows where Notes has a value
df.dropna(subset=["Notes"])
```

```python
#DROPPING
df_without_notes = df.drop(columns=["Notes"])
print("\nDataFrame without Notes column:\n", df_without_notes.head())
```

```python
#JOIN
df = pd.read_csv("Stocks_Tanjeem.csv", index_col="Symbol")
df_prices = pd.read_csv("Stock-Prices_Tanjeem.csv", index_col="Symbol") 
df_merged = df.join(df_prices, on="Symbol")
print("\nJoined DataFrame with Stock Prices:")
df_merged.head()
```
